```python
import re

# 유니코드 한글 시작 : 44032, 끝 : 55199
BASE_CODE, CHOSUNG, JUNGSUNG = 44032, 588, 28

# 초성 리스트. 00 ~ 18
CHOSUNG_LIST = ['ㄱ', 'ㄲ', 'ㄴ', 'ㄷ', 'ㄸ', 'ㄹ', 'ㅁ', 'ㅂ', 'ㅃ', 'ㅅ', 'ㅆ', 'ㅇ', 'ㅈ', 'ㅉ', 'ㅊ', 'ㅋ', 'ㅌ', 'ㅍ', 'ㅎ']

def sentence_to_chosung(sentence):
    split_keyword_list = list(sentence)
    result = list()
    for keyword in split_keyword_list:
        # 한글 여부 check 후 분리
        if re.match('.*[ㄱ-ㅎㅏ-ㅣ가-힣]+.*', keyword) is not None:
            try:
                char_code = ord(keyword) - BASE_CODE
                char1 = int(char_code / CHOSUNG)
                chosung = CHOSUNG_LIST[char1]
            except:
                chosung = keyword
            result.append(chosung)
        else:
            result.append(keyword)
    return "".join(result)
```

```python
sentence_to_chosung('안녕하세요 <number>곽근봉입니다.')
```

```python
sentence_file = open('/home/tf_user/test_data/namuwiki_sentences.txt','r')
chosung_output_file = open('/home/tf_user/test_data/namuwiki_chosung.txt','w')

line_counter = 0
while True:    
    line = sentence_file.readline()
    if not line: break
    line_counter += 1
    chosung_output_file.write(sentence_to_chosung(line))
        
    if line_counter % 1000 == 0:
        print('step:',line_counter)

sentence_file.close()
chosung_output_file.close()
```
